```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import linregress
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats

df1 = pd.read_csv("../..//..//Data//CleanData_mileage_price_enginesize.csv")

df1.head()
```

```python
conditions = [
    (df1['model'] == 'Yaris') | (df1['model'] == 'Aygo'),
    (df1['model'] == 'Fiesta') | (df1['model'] =='Focus')
    ]

# create a list of the values we want to assign for each condition
values = ['Toyota', 'Ford',]

# create a new column and use np.select to assign values to it using our lists as arguments
df1['make'] = np.select(conditions, values)

# display updated DataFrame
df1.head()
```

```python
color_dict2 = dict({'Toyota':'#EB0A1E',
                  'Ford':'#2A6BAC'})
ax = sns.violinplot(x="year", y="price", hue="make", split=True, data=df1, palette = color_dict2)
ax.set_title('Distribution of price', fontsize=16);

plt.gcf().set_size_inches(15,10)
plt.savefig("out.png") 
plt.show()
```
